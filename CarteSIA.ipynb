```python
import folium
import pandas as pd
```

```python
my_map = folium.Map(
    location=[47.17686, -2.29008]
)
my_map
```

```python
interv = pd.read_csv('CSV_SIA.csv')
interv.head(4)
```

```python
def select_marker_color(row) :
    if row['Etat Intervention'] == 'en cours' :
        return 'green'
    elif row['Etat Intervention'] == 'programmée' :
        return 'red'
    elif row['Etat Intervention'] == 'terminée' :
        return 'gray' 
    return 'blue' #l'intervention est en attente dans la file
```

```python
interv['color'] = interv.apply(select_marker_color, axis = 1)
interv.head(4)
```

```python
#inter = interv.loc[0]

for _, inter in interv.iterrows():
    folium.Marker(
        location = [inter['Latitude'], inter['Longitude']],
        popup = inter['Client'],
        tooltip = inter['Client'],
        icon = folium.Icon(color = inter['color'])
    ).add_to(my_map)

my_map
```

```python
def update_data(dataF, form_addon): #form_addon est un string de la forme "nomFichier.extension"
    fichier = open(form_addon, 'r').readlines()
    #on suppose que les infos du fichier sont dans le bon ordre 
    dico = {'N° commande': fichier[15], 
            'Date' : fichier[13],
            'Client' : fichier[1],
            'N° Client' : fichier[3],
            'Contact' : fichier[5],
            "Pays d'intervention" : fichier[7],
            'Latitude' : fichier[9],
            'Longitude' : fichier[10],
            'Etat Intervention' : 'en attente'
           }
    dataF = dataF.append(dico, ignore_index = True)
    return dataF
```

```python
interv = update_data(interv, 'LabsLittleMermaid.txt')
interv['color'] = interv.apply(select_marker_color, axis = 1)
```

```python
for _, inter in interv.iterrows():
    folium.Marker(
        location = [inter['Latitude'], inter['Longitude']],
        popup = inter['Client'],
        tooltip = inter['Client'],
        icon = folium.Icon(color = inter['color'])
    ).add_to(my_map)

my_map
```
